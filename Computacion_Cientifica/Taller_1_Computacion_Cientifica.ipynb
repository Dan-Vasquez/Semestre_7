```python

############################ 8
print("Octavo")

def mayor_mat(A):
  mayor = A[0,0]
  posicion = [[0,0]]
  [f,c] = np.shape(A)
  for i in range(0,f,1):
    for j in range(0,c,1):
      if mayor < A[i, j]:
        mayor = A[i,j]
        posicion = [[i, j]]
      elif mayor == A[i,j]:
        posicion.append([i,j])
  return mayor, posicion
A = np.array([[2,3,5],[-3,5,4]])
print(mayor_mat(A))

############################ 9

print("Noveno")

def es_matriz_probabilidad(A):
  ans = True

  # Verificar si todos los elementos son no negativos.
  if not np.all(A >= 0):
    ans = False
  # Verificar si la suma de los elementos en cada renglón es 1.
  for i in range(A.shape[0]):
    if np.sum(A[i, :]) != 1:
      ans = False

  return ans

A = np.array([[0.5, 0.25, 0.25], [0.3, 0.4, 0.3], [0.2, 0.3, 0.5]])

if es_matriz_probabilidad(A):
  print("La matriz es una matriz de probabilidad.")
else:
  print("La matriz no es una matriz de probabilidad.")
```
