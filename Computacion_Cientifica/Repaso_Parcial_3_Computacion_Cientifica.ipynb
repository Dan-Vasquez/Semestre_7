```python
def derivada(func, Xi, h):
    atras = (func(Xi) - func(Xi - h)) / h
    adelante = (func(Xi + h) - func(Xi)) / h
    central = (func(Xi + h) - func(Xi - h)) / (2 * h)
    return atras, adelante, central

def segunda_derivada(func, Xi, h):
    atras = (func(Xi) - 2 * func(Xi - h) + func(Xi - 2 * h)) / (h ** 2)
    adelante = (func(Xi + 2 * h) - 2 * func(Xi + h) + func(Xi)) / (h ** 2)
    central = (func(Xi + h) - 2 * func(Xi) + func(Xi - h)) / (h ** 2)
    return atras, adelante, central

def error(real, aprox):
  return np.abs(real-aprox)/np.abs(real)*100
```

```python
# Datos proporcionados
t = np.array([0, 25, 50, 75, 100, 125])
s = np.array([0, 32, 58, 78, 92, 100])

# Diferenciación numérica para velocidad y aceleración
def calcular_velocidad(t, s):
    n = len(t)
    v = np.zeros(n)

    # Velocidad hacia adelante en el primer punto
    v[0] = (s[1] - s[0]) / (t[1] - t[0])

    # Velocidad hacia atrás en el último punto
    v[-1] = (s[-1] - s[-2]) / (t[-1] - t[-2])

    # Velocidad centrada para los puntos intermedios
    for i in range(1, n-1):
        v[i] = (s[i+1] - s[i-1]) / (t[i+1] - t[i-1])

    return v

def calcular_aceleracion(t, s):
    n = len(t)
    a = np.zeros(n)

    # Aceleración hacia adelante en el primer punto
    a[0] = (s[2] - 2 * s[1] + s[0]) / ((t[1] - t[0]) ** 2)

    # Aceleración hacia atrás en el último punto
    a[-1] = (s[-1] - 2 * s[-2] + s[-3]) / ((t[-1] - t[-2]) ** 2)

    # Aceleración centrada para los puntos intermedios
    for i in range(1, n-1):
        a[i] = (s[i+1] - 2 * s[i] + s[i-1]) / ((t[i+1] - t[i]) ** 2)

    return a

# Calcular velocidad y aceleración
velocidad = calcular_velocidad(t, s)
aceleracion = calcular_aceleracion(t, s)

# Mostrar resultados
print("Tiempo (t):", t)
print("Distancia (s):", s)
print("Velocidad (v):", velocidad)
print("Aceleración (a):", aceleracion)
```

```python
# f funcion en el integrando
#[a,b] limites de integracion
# (n - 1 ) numero de particiones (n - numero de puntos)
def regla_trapecio(f,a,b,n):
  x = np.linspace(a,b,n)
  suma = 0
  for i in range(0, n-1,1):
    suma += (x[i+1]-x[i])/2*(f(x[i])+f(x[i+1]))
  return suma
```

```python
# f funcion en el integrando
#[a,b] limites de integracion
# (n - 1 ) numero de particiones (n - numero de puntos)
# n impar
def regla_sipmson(f,a,b,n):
  if n%2 != 0:
    x = np.linspace(a,b,n)
    suma = 0
    for i in range(0, int((n-1)/2),1):
      suma += (x[2*i+2]-x[2*i])/6*(f(x[2*i])+4*f(x[2*i+1])+f(x[2*i+2]))
    return suma
  else:
    print("n debe ser impar")
```

```python
def simpson(f, n, a, b):
    # Verificar si el número de divisiones es impar
    if n % 2 != 0:
        return tabla_trapecio(f, a, b, n)

    # Generar divisiones y verificar si están uniformemente espaciadas
    divisiones = np.linspace(a, b, n + 1)
    h = divisiones[1] - divisiones[0]
    for i in range(1, len(divisiones)):
        if not np.isclose(divisiones[i] - divisiones[i-1], h):
            return tabla_trapecio(f, a, b, n)

    # Si se cumplen las condiciones, aplicar la regla de Simpson
    sum = 0
    for i in range(0, len(divisiones) - 2, 2):
        sum += (divisiones[i + 2] - divisiones[i]) / 6 * (f(divisiones[i]) + 4 * f(divisiones[i + 1]) + f(divisiones[i + 2]))
    return sum
```

```python
def simpson_tabla(x, y):
  n = len(x)
  if n % 2 == 0:
    return "Error, entrada no valida, los numero de particiones no son impares. Por ende, se usara trapecios :", tabla_trapecio(x, y)
  else:
    su = 0
    for i in range(0, int((n - 1)/2), 1):
      su += (x[2 * i + 2] - x[2 * i]) / 6 * (y[2 * i] + 4 * y[2 * i + 1] + y[2 * i + 2])
    return su
```

```python
# f funcion funte, a, b limites de tiempo.
# n numero de particiones, ci condicion inicial

def metEuler(f, a, b, n, ci):
  t = np.linspace(a, b, n+1)
  print(t)
  y_apro = t*0
  y_apro[0] = ci
  h = (b-a)/n
  for i in range(n):
    y_apro[i+1] = y_apro[i] + h*f(t[i], y_apro[i])
  return t, y_apro

def metRungeKutter2(f, a, b, n, ci):
  t = np.linspace(a, b, n+1)
  y_apro = t*0
  y_apro[0] = ci
  h = (b-a)/n
  for i in range(n):
    k1 = f(t[i], y_apro[i]) # h * f(t[i], y_apro[i])
    k2 = f(t[i] + h, y_apro[i] + h*k1) # h * f(t[i] + h/2, y_apro[i] + k1 / 2)
    y_apro[i+1] = y_apro[i] + h*(k1 + k2)/2 # y_apo[i] + k2
  return t, y_apro

def metRungeKutter4(f, a, b, n, ci):
  t = np.linspace(a, b, n+1)
  y_apro = t*0
  y_apro[0] = ci
  h = (b-a)/n
  for i in range(n):
    k1 = f(t[i], y_apro[i])
    k2 = f(t[i] + h/2, y_apro[i] + h/2*k1)
    k3 = f(t[i] + h/2, y_apro[i] + h/2*k2)
    k4 = f(t[i+1], y_apro[i] + h*k3)
    y_apro[i+1] = y_apro[i] + h*(k1 + 2*k2 + 2*k3 + k4)/6
  return t, y_apro
```
