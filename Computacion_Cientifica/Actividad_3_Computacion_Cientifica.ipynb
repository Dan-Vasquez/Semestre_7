```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

# Funciones

```python
def modelo_sinusoidal(t, A, C, omega, teta):

    """
    A : valor medio
    C : amplitud
    omega : frecuencia angular
    teta : corrimiento de fase
    """

    y = A + C * np.cos(omega * t + teta)
    return y
```

```python
def sumatorias(t, y, omega):

  """
  t : lista de valores de tiempo
  y : lista de valores de la funcion del modelo sinusoidal
  omega : frecuencia angular
  """

  y_sum, ycos, ysen = 0, 0, 0

  for i in range(len(t)):
    y_sum += round(y[i], 2)
    ycos += round(round(y[i], 2) * np.cos(omega * t[i]), 3)
    ysen += round(round(y[i], 2)* np.sin(omega * t[i]), 3)

  return y_sum, ysen, ycos
```

```python
def coeficientes(t, y, ycos, ysen):
  """
  Funcion ajuste Sinusoide de minimos cuadrados

  t : lista de valores de tiempo
  y : sumatoria de la lista de valores de y
  ycos : sumatoria de la lista de valores de ycos
  ysen : sumatoria de la lista de valores de ysen
  """

  N = len(t)
  A0 = round(y/N, 2)
  A1 = round((2/N) * ycos, 3)
  B1 = round((2/N) * ysen, 3)
  return A0, A1, B1
```

```python
def ajuste_por_minimos_cuadrados(t, A0, A1, B1, omega):
  """
  t : lista de valores de tiempo
  A0, A1, B1 : coeficientes
  omega : frecuencia angular
  """

  y = A0 + A1 * np.cos(omega * t) + B1 * np.sin(omega * t)
  return y
```

```python
def graficar(t, y, y_ajuste):
    """
    Grafica los datos originales y el ajuste.
    """
    plt.figure(figsize=(5, 5))
    plt.plot(t, y, 'o', label='Datos originales')
    plt.plot(t, y_ajuste, 'r', label='Ajuste')
    plt.legend()
    plt.show()
```

# Prueba de la explicacion

```python
# Parmetros de los datos
A = 1.7
C = 1
omega = 4.189
teta = 1.0472

# Generar datos
t = np.linspace(0, 1.35, 10)

y_real = modelo_sinusoidal(t, A, C, omega, teta)

# Calcular parmetros del ajuste

y_sum, ysen, ycos = sumatorias(t, y_real, omega)

# Calcular coeficientes
A0, A1, B1 = coeficientes(t, y_sum, ycos, ysen)

# Ajuste por mnimos cuadrados
y_ajuste = ajuste_por_minimos_cuadrados(t, A0, A1, B1, omega)

# Graficar
graficar(t, y_real, y_ajuste)
```
